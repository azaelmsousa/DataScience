## REST search on SNOMED CT Ontology for resources which contains "Dengue" (case insensitive) within its label

### PS: REST search was performed since the ontology requires a complex SPARQL query to count resources because it uses different properties prefixes 

```python
import urllib.request, urllib.error, urllib.parse
import json
import os

from pprint import pprint

REST_URL = "http://data.bioontology.org"
API_KEY = "1ed557d6-a913-44c7-8097-10fedbf1b72d"

def get_json(url):
    opener = urllib.request.build_opener()
    opener.addheaders = [('Authorization', 'apikey token=' + API_KEY)]
    return json.loads(opener.open(url).read())

# Do a search for 'dengue' in SNOMED CT Ontology
term = "dengue"
ontologies = "SNOMEDCT"

search_results = get_json(REST_URL + "/search?q=" + term + "&ontologies=" + ontologies)["collection"]

print('RESOURCES FOUND: %d' % len(search_results))
```

## Show results as JSON

```python
# Print the results
for result in search_results:
    pprint(result)
```
